```python
import pandas as pd


train_file_path = "/home/yudai/Project/research/Vul_Detection/data/raw/new_six_by_projects/train.jsonl"
valid_file_path = "/home/yudai/Project/research/Vul_Detection/data/raw/new_six_by_projects/valid.jsonl"
test_file_path  = "/home/yudai/Project/research/Vul_Detection/data/raw/new_six_by_projects/test.jsonl"

def load_jsonl(path):
    """1 行 1 JSON の .jsonl を pandas DataFrame に読み込む"""
    return pd.read_json(path, lines=True)

# ----- ① load -----
df_train = load_jsonl(train_file_path)
df_valid = load_jsonl(valid_file_path)
df_test  = load_jsonl(test_file_path)

# ----- ② データ概要 -----
print("=== SHAPES ===")
print(f"train: {df_train.shape}")
print(f"valid: {df_valid.shape}")
print(f"test : {df_test.shape}")

print("\n=== COLUMNS ===")
print("train:", list(df_train.columns))
print("valid:", list(df_valid.columns))
print("test :", list(df_test.columns))

# ----- ③ 先頭の数件を表示 -----
print("\n=== TRAIN HEAD ===")
display(df_train.head(3))

print("\n=== VALID HEAD ===")
display(df_valid.head(3))

print("\n=== TEST HEAD ===")
display(df_test.head(3))

# ----- ④ ラベル分布やプロジェクト分布も確認（任意） -----
if "label" in df_train.columns:
    print("\n=== LABEL DISTRIBUTION (train) ===")
    print(df_train["label"].value_counts())

if "project" in df_train.columns:
    print("\n=== PROJECT COUNT (train) ===")
    print(df_train["project"].value_counts())
```

```python
# ----- ⑤ project カラムの種類一覧 -----

def show_project_stats(df, name):
    print(f"\n=== PROJECTS in {name} ({df.shape[0]} samples) ===")
    projects = sorted(df["project"].unique())
    print("count:", len(projects))
    print(projects)

    print(f"\n=== TOP PROJECT COUNTS in {name} ===")
    print(df["project"].value_counts().head(20))  # 上位20件を表示（必要なら増やせる）

# train / valid / test それぞれ表示
show_project_stats(df_train, "TRAIN")
show_project_stats(df_valid, "VALID")
show_project_stats(df_test, "TEST")

```
